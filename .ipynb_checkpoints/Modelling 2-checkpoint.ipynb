#### Make a baseline prediction of the 2019-2020 season



```python
df_2019 = pd.read_csv('data\game_results_2019.csv')
```

```python
df_2019.head()
```

```python
# In data exploration, found that the home team wins 55% of the time. Predict that the team with more points will win the game. If tied in points, predict team with more wins will win.
# If tied again, 55% home team will win.
df_2019.loc[(df_2019['home_points'] > df_2019['away_points']), 'prediction'] = 1
df_2019.loc[(df_2019['home_points'] == df_2019['away_points']) & (df_2019['home_wins'] > df_2019['away_wins']), 'prediction'] = 1
df_2019.loc[(df_2019['home_points'] == df_2019['away_points']) & (df_2019['home_wins'] < df_2019['away_wins']), 'prediction'] = 0
df_2019.loc[(df_2019['home_points'] == df_2019['away_points']) & (df_2019['home_wins'] == df_2019['away_wins']), 'prediction'] = (1 if random() > 0.55 else 0)
df_2019.loc[(df_2019['home_points'] < df_2019['away_points']), 'prediction'] = 0
df_2019['prediction'] = df_2019['prediction'].astype(int)
```

```python
df_2019.head()
```

```python
print("Accuracy:",accuracy_score(df_2019['winner'],df_2019['prediction']))
print("Precision:",precision_score(df_2019['winner'],df_2019['prediction'],average='macro'))
print("Recall:",recall_score(df_2019['winner'],df_2019['prediction'],average='macro'))
print("F1:",f1_score(df_2019['winner'],df_2019['prediction'],average='macro'))
```

```python
cnf_matrix_baseline = confusion_matrix(df_2019['winner'],df_2019['prediction'])
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix_baseline), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

#### Building Pipeline and different models

```python
df = pd.read_csv('data\game_results_2011_2018.csv')
```

```python
df.head()
```

```python
# Drop a couple of columns that wont be needed for making a prediction. Drop period, game date, home_games_played, away_games_played, home_points, away_points, home_score, away_score
df.drop(['period','game_date','home_games_played','away_games_played','home_points','away_points','home_score','away_score'],axis=1,inplace=True)
```

```python
# x is the home team, y is the away team
df.head()
```

```python
# Split data to train and test
X = df.drop('winner',axis=1)
y = df['winner']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=60)
```

```python
X_train
```

```python
cat_feats = X_train.dtypes[X_train.dtypes == 'object'].index.tolist()
num_feats = X_train.dtypes[~X_train.dtypes.index.isin(cat_feats)].index.tolist()
```

```python
num_pipe = make_pipeline(
    StandardScaler(),
    PCA(n_components=3)
)
cat_pipe = make_pipeline(
    OneHotEncoder()
)
```

#### XGB

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('continuous', num_pipe, num_feats),
        ('categorical', cat_pipe, cat_feats),
    ]
)

model_pipe = imblearn.pipeline.Pipeline(steps = [
    ('preprocess', preprocessor),
    ('classifier', XGBClassifier(objective='reg:logistic',colsample_bytree = 0.3, learning_rate = 0.1, max_depth=9,alpha=10,n_estimators=250))
])

param_grid = {'classifier__learning_rate': [0.001,0.01,0.1], 
              'classifier__n_estimators': [50,100,150,200,250],
              'preprocess__continuous__pca__n_components': [3,5,7,11],
             }

grid = GridSearchCV(model_pipe, param_grid=param_grid, cv=5, n_jobs=-1,verbose=1)
grid.fit(X_train, y_train)

best_model = grid.best_estimator_
best_hyperparams = grid.best_params_
best_acc = grid.score(X_test, y_test)
print(f'Best test set accuracy: {best_acc}\nAchieved with hyperparameters: {best_hyperparams}')
```

```python
grid_pred = grid.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test,grid_pred)
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots(figsize = (10,8))
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
