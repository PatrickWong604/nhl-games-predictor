```python
import pandas as pd
import requests
pd.set_option('display.max_columns',None)
import datetime as dt
from datetime import datetime
from time import sleep
from random import randrange
```

```python
range_2019 = pd.date_range(start='2019-10-02', end='2020-03-11')
range_2018 = pd.date_range(start='2018-10-03', end='2019-04-06')
range_2017 = pd.date_range(start='2017-10-04', end='2018-04-08')
range_2016 = pd.date_range(start='2016-10-12', end='2017-04-09')
```

```python
def strip_time(dates):
    dates = dates.to_pydatetime()
    date_list = []
    for i in range(len(dates)):
        temp_date = dates[i].strftime('%Y-%m-%d')
        date_list.append(temp_date)
    return date_list
```

```python
dates_2019 = strip_time(range_2019)
dates_2018 = strip_time(range_2018)
dates_2017 = strip_time(range_2017)
dates_2016 = strip_time(range_2016)
```
