```python
# need to fix the records columns, as they include the current game's results
def correct_records(df):
    # new home wins column
    df.loc[df['home_score'] > df['away_score'], 'home_wins_corr'] = (df['home_wins'] - 1)
    df.loc[df['home_score'] < df['away_score'], 'home_wins_corr'] = df['home_wins']
    # new home losses column
    df.loc[(df['home_score'] > df['away_score']), 'home_losses_corr'] = df['home_losses']
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == '3rd'), 'home_losses_corr'] = (df['home_losses']-1)
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == 'OT'), 'home_losses_corr'] = (df['home_losses'])
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == 'SO'), 'home_losses_corr'] = (df['home_losses'])
    # new home otl column
    df.loc[(df['home_score'] > df['away_score']), 'home_otl_corr'] = df['home_otl']
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == '3rd'), 'home_otl_corr'] = (df['home_otl'])
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == 'OT'), 'home_otl_corr'] = (df['home_otl']-1)
    df.loc[(df['home_score'] < df['away_score']) & (df['period'] == 'SO'), 'home_otl_corr'] = (df['home_otl']-1)  
    # new away wins column
    df.loc[df['away_score'] > df['home_score'], 'away_wins_corr'] = (df['away_wins'] - 1)
    df.loc[df['away_score'] < df['home_score'], 'away_wins_corr'] = df['away_wins']
    # new away losses column
    df.loc[(df['away_score'] > df['home_score']), 'away_losses_corr'] = df['away_losses']
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == '3rd'), 'away_losses_corr'] = (df['away_losses']-1)
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == 'OT'), 'away_losses_corr'] = (df['away_losses'])
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == 'SO'), 'away_losses_corr'] = (df['away_losses'])
    # new away otl column
    df.loc[(df['away_score'] > df['home_score']), 'away_otl_corr'] = df['away_otl']
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == '3rd'), 'away_otl_corr'] = (df['away_otl'])
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == 'OT'), 'away_otl_corr'] = (df['away_otl']-1)
    df.loc[(df['away_score'] < df['home_score']) & (df['period'] == 'SO'), 'away_otl_corr'] = (df['away_otl']-1)     
    df['home_wins_corr'] = df['home_wins_corr'].astype(int)
    df['home_losses_corr'] = df['home_losses_corr'].astype(int)
    df['home_otl_corr'] = df['home_otl_corr'].astype(int)
    df['away_wins_corr'] = df['away_wins_corr'].astype(int)
    df['away_losses_corr'] = df['away_losses_corr'].astype(int)
    df['away_otl_corr'] = df['away_otl_corr'].astype(int)
    df = df.drop(['home_wins','home_losses','home_otl','away_wins','away_losses','away_otl'], axis = 1)
    df.rename(columns={'home_wins_corr':'home_wins',
                    'home_losses_corr':'home_losses',
                    'home_otl_corr':'home_otl',
                    'away_wins_corr':'away_wins',
                    'away_losses_corr':'away_losses',
                    'away_otl_corr':'away_otl'}, inplace=True)
    return df  
```

```python
# short function to map the winner of the game
def winner(df):
    df['winner'] = np.where((df['home_score'] > df['away_score']), 1, 0)
    return df
```

```python
# short function to calculate the number of games played for each team
def games_played(df):
    df['home_games_played'] = df['home_wins'] + df['home_losses'] + df['home_otl']
    df['away_games_played'] = df['away_wins'] + df['away_losses'] + df['away_otl']
    return df
```

```python
# change some team names as the two data sources are a bit different
def generalize_name(df):
    df.loc[df['home_team'] == 'Montréal Canadiens','home_team'] = 'Montreal Canadiens'
    df.loc[df['away_team'] == 'Montréal Canadiens','away_team'] = 'Montreal Canadiens'
    df.loc[df['home_team'] == 'St. Louis Blues','home_team'] = 'St Louis Blues'
    df.loc[df['away_team'] == 'St. Louis Blues','away_team'] = 'St Louis Blues'
    df.loc[df['home_team'] == 'Phoenix Coyotes','home_team'] = 'Arizona Coyotes'
    df.loc[df['away_team'] == 'Phoenix Coyotes','away_team'] = 'Arizona Coyotes'
    return df
```

Corsi - Any shot attempt (goals, shots on net, misses and blocks) outside of the shootout. Referred to as SAT by the NHL <br>
CF% - Percentage of total Corsi in games that team played that are for that team. CF*100/(CF+CA)<br>

Fenwick - Any unblocked shot attempt (goals, shots on net and misses) outside of the shootout. Referred to as USAT by the NHL.<br>
FF% - Percentage of total Fenwick in games that team played that are for that team. FF*100/(FF+FA)<br>

Shots - Any shot attempt on net (goals and shots on net) outside of the shootout.<br>
SF% - Percentage of total Shots in games that team played that are for that team. SF*100/(SF+SA)<br>

Goals - Any goal, outside of the shootout.<br>
GF% - Percentage of total Goals in games that team played that are for that team. GF*100/(GF+GA)<br>

Scoring Chances - A scoring chance, as originally defined by War-on-Ice.<br>
Each shot attempt (Corsi) taken in the offensive zone is assigned a value based on the area of the zone in which it was recorded. Attempts made from the attacking team's neutral or defensive zones are excluded.<br>
SCF% - Percentage of total Scoring Chances in games that team played that are for that team. SCF*100/(SCF+SCA)<br>

Scoring Chance Shots - Any shot attempt on net (goals and shots on net) that qualifies as a Scoring Chance.<br>
SCSF% - Percentage of total Shots that are scoring chances in games that team played that are for that team. SCSF*100/(SCSF+SCSA)<br>

Scoring Chance Goals - Goals generated from scoring chances<br>
SCSH% - Percentage of Scoring Chance Shots for that team that were Goals. SCGF*100/SCSF<br>
SCSV% - Percentage of Scoring Chance Shots against that team that were not Goals. 100-(SCGA*100/SCSA)<br>

High Danger Scoring Chances - A scoring chance with a score of 3 or higher.<br>
HDCF% - Percentage of total High Danger Scoring Chances in games that team played that are for that team. HDCF*100/(HDCF+HDCA)<br>

High Danger Shots - Any shot attempt on net (goals and shots on net) that qualifies as a High Danger Scoring Chance.<br>
HDSF% - Percentage of total Shots that are High Danger Scoring Chances in games that team played that are for that team. HDSF*100/(HDSF+HDSA)<br>

High Danger Goals - Goals generated from High Danger Scoring Chances<br>
HDGF% - Percentage of total Goals off of High Danger Scoring Chances in games that team played that are for that team. HDGF*100/(HDGF+HDGA)<br>
HDSH% - Percentage of High Danger Shots for that team that were Goals. HDGF*100/HDSF<br>
HDSV% - Percentage of High Danger Shots against that team that were not Goals. 100-(HDGA*100/HDSA)<br>

Medium Danger Scoring Chances - A scoring chance with a score of exactly 2.<br>
MDCF% - Percentage of total Medium Danger Scoring Chances in games that team played that are for that team. MDCF*100/(MDCF+MDCA)<br>

Medium Danger Shots - Any shot attempt on net (goals and shots on net) that qualifies as a Medium Danger Scoring Chance.<br>
MDSF% - Percentage of total Shots that are Medium Danger Scoring Chances in games that team played that are for that team. MDSF*100/(MDSF+MDSA)<br>

Medium Danger Goals - Goals generated from Medium Danger Scoring Chances<br>
MDGF% - Percentage of total Goals off of Medium Danger Scoring Chances in games that team played that are for that team. MDGF*100/(MDGF+MDGA)<br>
MDSH% - Percentage of Medium Danger Shots for that team that were Goals. MDGF*100/MDSF<br>
MDSV% - Percentage of Medium Danger Shots against that team that were not Goals. 100-(MDGA*100/MDSA)<br>

Low Danger Scoring Chances - A scoring chance with a score of 1 or less. Does not include any attempts from the attacking team's neutral or defensive zone.<br>
LDCF% - Percentage of total Low Danger Scoring Chances in games that team played that are for that team. LDCF*100/(LDCF+LDCA)<br>

Low Danger Shots - Any shot attempt on net (goals and shots on net) that qualifies as a Low Danger Scoring Chance.<br>
LDSF% - Percentage of total Shots that are Low Danger Scoring Chances in games that team played that are for that team. LDSF*100/(LDSF+LDSA)<br>

Low Danger Goals - Goals generated from Low Danger Scoring Chances<br>
LDSH% - Percentage of Low Danger Shots for that team that were Goals. LDGF*100/LDSF<br>
LDSV% - Percentage of Low Danger Shots against that team that were not Goals. 100-(LDGA*100/LDSA)<br>

PDO
SH% - Percentage of Shots for that team that were Goals. GF*100/SF<br>
SV% - Percentage of Shots against that team that were not Goals. 100-(GA*100/SA)<br>
PDO - Shooting percentage plus save percentage. (GF/SF)+(GA/SA)<br>
